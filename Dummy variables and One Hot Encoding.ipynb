```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv(r'G:\Machine Learning\jupyter csv files\One Hot Coding TC.csv') #to upload a file from differentlocation use r before quotes
```

```python
df
```

```python
#by using dummy variables 
```

```python
pd.get_dummies(df.Township)   #create binary codes for the column we mention(Township)
```

```python
dv = pd.get_dummies(df.Township)  #creating the binary code in the name of a dummmy variable(dv)
dv
```

```python
merge = pd.concat([df,dv], axis = 'columns') #concartinating the data frame-file(df) with dummy variable in the column manner specified by column 
```

```python
merge
```

```python
final = merge.drop(['Township','Porur'], axis='columns')  #drop the township column as we no longer need it
                      # drop one of the dummy variable column as it may creat dummy variable trap(if there are 5 column drop 1 and use 4 and so on)
final        #sklearn knows about dummy variable trap and its not necessary to remove but in general its good to drop a dummy variable
```

```python
X = final.drop(['Price'],axis='columns')
X
```

```python
Y = final.Price
```

```python
Y
```

```python
lr = linear_model.LinearRegression()
lr
```

```python
lr.fit(X,Y)
```

```python
lr.predict([[2800,0,1]])  #predicting value for KK nagar, check X and write the value, 2800-area 0-guindy 1-KKnagar 
```

```python
lr.predict([[3400,0,0]])  #predicting value for porur, since we deleted that write 00 for it  
```

```python
lr.score(X,Y)  #as by LR the values of m and b in y=mx+b is calculated and thus X is calculated 
                #predicits the value for data in X and compare with Y and do some calculation to sa how accurate the model is 
                #is score is 1 the model is perfect
```

```python
from sklearn.preprocessing import LabelEncoder #to use hot encoder we need to do lable encoding on town column 
le = LabelEncoder()   #creating a class object for Lable encoder  
```

```python
le.fit_transform(dfle.Township)  #fit_transform takes my lable column as input and return their lables
```

```python
X= dfle[['Township','Area']].values    #.values convert data frame to array 
X
```

```python
y = dfle.Price
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
```

```python
X
```

```python
X = X[:,1:]     #to avoid dummy variable trapping slice a column (column 0 from the array)
X 
```
