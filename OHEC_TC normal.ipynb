```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv(r'G:\Machine Learning\jupyter csv files\OHC TC.csv') #to upload a file from differentlocation use r before quotes
```

```python
df
```

```python
dv = pd.get_dummies(df.Car_Model)   #create binary codes for the column we mention(Township)
dv
```

```python
merge = pd.concat([dv,df], axis = 'columns')
```

```python
merge
```

```python
final = merge.drop(["Car_Model"],axis = 'columns')
```

```python
final
```

```python
lr = linear_model.LinearRegression()
lr
```

```python
lr.fit(final[['Mileage','Age','Audi A5','BMW X5','Mercedez Benz C Class']],final.Sell_Price)
```

```python
lr.predict([[45000,4,0,0,1]])
```

```python
lr.score(final[['Mileage','Age','Audi A5','BMW X5','Mercedez Benz C Class']],final.Sell_Price)
```
